```
import matplotlib.pyplot as plt
```

```
apples=[0.895,0.91,0.919,0.926,0.920,0.931]
years=[2013,2014,2015,2016,2017,2018]
plt.plot(years,apples)
```

```
plt.plot(years,apples)
plt.xlabel("years")
plt.ylabel;("yield(tons per hectar)")
```

```
year=range(2013,2019)
apples=[0.895,0.91,0.919,0.926,0.920,0.931]
oranges=[0.962,0.955,0.954,0.952,0.949,0.945]
plt.plot(year,apples)
plt.plot(year,oranges)
plt.xlabel("years")
plt.ylabel("yield")
plt.legend(["apples","oranges"])
```

```
plt.plot(year,apples,'s-b',year,oranges,'o-r')
plt.xlabel("years")
plt.ylabel("yield")

plt.xlabel("years")
plt.ylabel("yield")

plt.title('crop yield in kanto')
plt.legend(["apples","oranges"])
```

```
plt.figure(figsize=(4,4))

plt.plot(year,oranges,'or')
plt.title('yield of oranges')
```
